**Q** A dataframe month wise ?

```python
import datetime
```

```python
Covid_MAR1['date'] = pd.to_datetime(Covid_MAR1['date'])
Covid_MAR1['Year/Month'] = Covid_MAR1['date'].apply(lambda x: x.strftime('%Y')) + ' ' + Covid_MAR1['date'].apply(lambda x: x.strftime('%B'))
Covid_MAR1
```

```python
Covid_MAR1_montly = Covid_MAR1.groupby('Year/Month')[['new_cases','new_deaths']].sum()
Covid_MAR1_montly['Death %'] = (Covid_MAR1_montly.new_deaths / Covid_MAR1_montly.new_cases)*100
Covid_MAR1_montly
```

**Q** When did the death % hit the max :

```python
Covid_MAR1_montly[Covid_MAR1_montly['Death %'] == max(Covid_MAR1_montly['Death %'])]
```

**Q** Data Year wise 


```python
Covid_MAR1['date'] = pd.to_datetime(Covid_MAR1['date'])
Covid_MAR1['Year'] = Covid_MAR1['date'].apply(lambda x: x.strftime('%Y'))
Covid_MAR1_Yearly = Covid_MAR1.groupby('Year')[['new_cases', 'new_deaths']].sum()
Covid_MAR1_Yearly['Death %'] = (Covid_MAR1_Yearly.new_deaths / Covid_MAR1_Yearly.new_cases)*100

Covid_MAR1_Yearly
```

**Q** Group the dataframe by months 


```python
deaths_df['date'] = pd.to_datetime(deaths_df['date'])
deaths_df[['date']]
```

```python
deaths_df['Year'] = pd.DatetimeIndex(deaths_df['date']).year
deaths_df['Month'] = pd.DatetimeIndex(deaths_df['date']).month
deaths_df['Day'] = pd.DatetimeIndex(deaths_df['date']).day

deaths_df
```

```python
deaths_grouped_df = deaths_df.groupby(['Year','location','Month'])[['new_cases', 'new_deaths']].sum()

deaths_grouped_df
```

**Q** let's break things down by continent


```python
Covid_continent_df = deaths_df.groupby('continent')[['new_cases', 'new_deaths']].sum()
Covid_continent_df
```

```python
Covid_continent_df['Death %'] = (Covid_continent_df['new_deaths']/Covid_continent_df['new_cases'])*100
Covid_continent_df['Population'] = deaths_df.groupby('continent')['population'].unique().apply(sum)
Covid_continent_df['Average $R_0$'] = deaths_df.groupby('continent')['reproduction_rate'].mean()
Covid_continent_df
```

```python
#Deleting the column 'Average Reproduction Rate'
Covid_continent_df.drop('Average Reproduction Rate', axis =1, inplace=True)
```
