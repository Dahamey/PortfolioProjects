**Q** Movies Vs TvShows in recent years :


```python
netflix_df.release_year.unique()
```

```python
tvShows_vs_Movies = netflix_df.groupby(['release_year','type'], as_index=False)[['title']].count()
tvShows_vs_Movies
```

Quelle est la tendance de ces 5 dernières années ?

```python
tvShows_vs_Movies = netflix_df.groupby(['release_year','type'])[['title']].count()
#.index.get_level_values(0) veut dire la 1ère groupby
tvShows_vs_Movies.loc[tvShows_vs_Movies.index.get_level_values(0) >= tvShows_vs_Movies.index.get_level_values(0).max() - 4]
```

Il y a eu une certaine focalisation sur les séries TV au cours des 5 dernières années
